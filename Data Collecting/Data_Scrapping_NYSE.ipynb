```python
import requests
import bs4
import pandas as pd
import json
import numpy
```

```python
payloadHeader = {
    "Content-Type": "application/json",
}

payloadData = {
    "instrumentType":"EQUITY",
    "pageNumber":1,
    "sortColumn":"NORMALIZED_TICKER",
    "sortOrder":"ASC",
    "maxResultsPerPage":1000000,
    "filterToken":"",
}
```

```python
url = "https://www.nyse.com/api/quotes/filter"
html = requests.post(url, data=json.dumps(payloadData), headers=payloadHeader)
listing = json.loads(html.text)
```

```python
fileObject = open('listing.json', 'w')
fileObject.write(html.text)
fileObject.close()
```

```python
company = list()
for i in range(6446):
    ticker = listing[i]["symbolTicker"].replace('.', '').split("-")[0]
    name = listing[i]["instrumentName"]
    url = f"https://finance.yahoo.com/quote/{ticker}/profile"
    html = requests.get(url)
    response = bs4.BeautifulSoup(html.content,"lxml")
    print(f"{i} {ticker}")
    try:
        website = response.find("div", class_="Mb(25px)").find_all("a")[1].get_text()
        sector = response.find("div", class_="Mb(25px)").find_all("span",class_="Fw(600)")[0].get_text()
        industry = response.find("div", class_="Mb(25px)").find_all("span",class_="Fw(600)")[1].get_text()
        description = response.find("p",class_="Mt(15px) Lh(1.6)").get_text()
        company.append((ticker, name, website, sector, industry, description))
    except:
        print("fail")
```

```python
listing = pd.DataFrame(company, columns=['ticker','name','website','sector', 'industry', 'description']).drop_duplicates('website')
listing = listing.reset_index(drop=True)
listing.to_csv("listing.csv")
```
