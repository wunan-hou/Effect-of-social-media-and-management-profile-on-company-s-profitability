```python
import requests
import bs4
import tweepy
import pandas as pd
import numpy as np
import yfinance as yf
import time
```

```python
# read twitter api
with open(r"tokens.txt",'r') as token_file:
    contents = token_file.read().split('\n')
    consumer_key = contents[0]
    consumer_secret = contents[1]
    access_token = contents[2]
    access_token_secret = contents[3]
```

```python
dataset = pd.read_csv("basic_data.csv")
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
twitter_list = list()
for html in list(dataset['Twitter'])[0:20]:
    try:
        screen_name = html.split("/")[-1]
        result = api.get_user(screen_name)
        name = result.screen_name
        date = result.created_at
        description = result.description
        followers = result.followers_count
        friends = result.friends_count
        favourites = result.favourites_count
        tweets = result.statuses_count
        verified = result.verified
        tweets_20 = api.user_timeline(screen_name)
        retweet = 0
        favorite = 0
        for i in range(20):
            retweet += tweets_20[i].retweet_count
            favorite += tweets_20[i].favorite_count
        avg_retweet = retweet/20
        avg_favorite = favorite/20
        twitter_list.append((name,date,description,followers,friends,favourites,tweets,verified,avg_retweet,avg_favorite))
    except:
        pass
```

```python
twitter_list
```

```python
twitter_df = pd.DataFrame(twitter_list,columns=['company','name','create_date','description','followers','friends','favourites','tweets','verified','avg_retweet','avg_favorite'])
```

```python
twitter_df.to_csv("sample_data_twitter.csv")
```

```python
m_list = list()
```

```python
dataset = pd.read_csv("data_v6.csv")
tickers = list(dataset['ticker'])
j = 2513
while True:
    for ticker in tickers[j:]:
        url = f"https://finance.yahoo.com/quote/{ticker}/financials"
        html = requests.get(url)
        results = bs4.BeautifulSoup(html.content,"lxml")
        try:
            result = results.find("div", class_="M(0) Mb(10px) Whs(n) BdEnd Bdc($seperatorColor) D(itb)")
            table = result.find_all("span")
            for i in range(len(table)):
                if table[i].get_text() == "Breakdown":
                    date_1 = table[i+2].get_text()
                    date_2 = table[i+3].get_text()
                    date_3 = table[i+4].get_text()
                if table[i].get_text() == "Total Revenue":
                    revenue_1 = table[i+2].get_text()
                    revenue_2 = table[i+3].get_text()
                    revenue_3 = table[i+4].get_text()
                if table[i].get_text() == "Operating Income or Loss":
                    opincome_1 = table[i+2].get_text()
                    opincome_2 = table[i+3].get_text()
                    opincome_3 = table[i+4].get_text()
                if table[i].get_text() == "Net Income":
                    netincome_1 = table[i+2].get_text()
                    netincome_2 = table[i+3].get_text()
                    netincome_3 = table[i+4].get_text()
            m_list.append((ticker,date_1,date_2,date_3,revenue_1,revenue_2,revenue_3,opincome_1,opincome_2,opincome_3,netincome_1,netincome_2,netincome_3))
            print(f"{j} {ticker}")
            j += 1
        except:
            break
    time.sleep(30)
```

```python
data = pd.DataFrame(m_list,columns=['ticker','date_1','date_2','date_3','revenue_1','revenue_2','revenue_3','opincome_1','opincome_2','opincome_3','netincome_1','netincome_2','netincome_3'])
```

```python
data.to_csv("revenue.csv")
```
