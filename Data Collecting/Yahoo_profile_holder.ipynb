```python
import requests
import bs4
import pandas as pd
import json
import numpy as np
```

```python
payloadHeader = {
    "Content-Type": "application/json",
}

payloadData = {
    "instrumentType":"EQUITY",
    "pageNumber":1,
    "sortColumn":"NORMALIZED_TICKER",
    "sortOrder":"ASC",
    "maxResultsPerPage":1000000,
    "filterToken":"",
}
```

```python
url = "https://www.nyse.com/api/quotes/filter"
html = requests.post(url, data=json.dumps(payloadData), headers=payloadHeader)
listing = json.loads(html.text)
```

```python
print(listing)
```

```python
fileObject = open('listing.json', 'w')
fileObject.write(html.text)
fileObject.close()
```

```python
def get_loc(string):
    import re
    pattern = r'([\w ]+,)+(?P<states>[a-zA-Z ]+)(?P<zipcode>[0-9]{2}[0-9]+)(?P<country>[a-zA-Z ]+)[0-9A-Za-z:/. ]*'
    match = re.search(pattern,string)
    states = match.group('states').strip()
    zipcode = match.group('zipcode').strip()
    country = match.group('country').strip()
    return states,zipcode,country
```

```python
def get_exec(response):
    executives = response.find('table',class_ = 'W(100%)').find_all('tr',class_ = 'C($primaryColor) BdB Bdc($seperatorColor) H(36px)')
    exec_list = []
    for i in range(5):
        temp = executives[i]
        name = temp.find('td',class_ = 'Ta(start)').get_text().strip()
        if name[:2] == 'Mr':
            gender = 'M'
        elif name[:2] == 'Ms' or name[:3] == 'Mrs':
            gender = 'F'
        else:
            gender = 'Na'
        title = temp.find('td',class_ = 'Ta(start) W(45%)').get_text().strip()
        birth_year = temp.find_all('td',class_ = 'Ta(end)')[-1].get_text().strip()
        if birth_year == 'N/A':
            age = 'N'
        #'N' means Na
        else:
            import datetime
            current_year = int(datetime.datetime.now().year)
            age = current_year - int(birth_year)
        exec_list.append(str(', '.join(str(i) for i in [gender,title,age])))
    return exec_list
    
```

```python
def get_profile():
    profile = list()
    for i in range(6446):
        ticker = listing[i]["symbolTicker"].replace('.', '').split("-")[0]
        name = listing[i]["instrumentName"]
        url = f"https://finance.yahoo.com/quote/{ticker}/profile"
        html = requests.get(url)
        response = bs4.BeautifulSoup(html.content,"lxml")
        try:
            address = response.find('p',class_='D(ib) W(47.727%) Pend(40px)').get_text().strip()
            states,zipcode,country = get_loc(address)
            employee = response.find("div", class_="Mb(25px)").find_all("span",class_="Fw(600)")[2].get_text()
            if employee == '':
                employee = None
            else:
                employee = int(''.join('1200'.split(',')))
            [exec1,exec2,exec3,exec4,exec5] = get_exec(response)
            profile.append((ticker,states,zipcode,country,employee,exec1,exec2,exec3,exec4,exec5))
        except:
            print("fail")
    return profile
```

```python
def get_holders():
    holder = list()
    j = 0
    for i in range(6446):
        ticker = listing[i]["symbolTicker"].replace('.', '').split("-")[0]
        name = listing[i]["instrumentName"]
        url = f"https://finance.yahoo.com/quote/{ticker}/holders"
        html = requests.get(url)
        response = bs4.BeautifulSoup(html.content,"lxml")
        try:
            table = response.find('table',class_ = 'W(100%) M(0) BdB Bdc($seperatorColor)')
            breakdown = table.find_all('tr',class_ = 'BdT Bdc($seperatorColor)')
            all_insider = breakdown[0].find('td',class_='Py(10px) Va(m) Fw(600) W(15%)').get_text().strip()
            share_by_inst = breakdown[1].find('td',class_='Py(10px) Va(m) Fw(600) W(15%)').get_text().strip()
            float_by_inst = breakdown[2].find('td',class_='Py(10px) Va(m) Fw(600) W(15%)').get_text().strip()
            num_inst = breakdown[3].find('td',class_='Py(10px) Va(m) Fw(600) W(15%)').get_text().strip()
            holder.append((ticker,all_insider,share_by_inst,float_by_inst,num_inst))
        except:
            j = j+1
            print(j,"fail")
    return holder
```
