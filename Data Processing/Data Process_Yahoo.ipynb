```python
for i in sectors:
    if i == np.nan:
        pass
    
    df_subset = df[df['Sector'] == i][['number_of_full-time_employee','Shares_Held_by_All_Insider','Shares_Held_by_Institutions',
                                      'Float_Held_by_Institutions','Number_of_Institutions_Holding_Shares','male_rate',
                                      'average_age','age_below_40','tech_score','margin_change']]
    temp_list = []
    for j in df_subset.columns:
        coverage = 1-df_subset[j].isnull().sum()/len(df_subset)
        min_value = df_subset[j].describe()['min']
        percentile = df_subset[j].quantile([.01, .05, .10, .25, .5,.7,.75,.8,.85, .90, .95, .99])
        p1 = percentile[0.01]
        p5 = percentile[0.05]
        p10 = percentile[0.10]
        p25 = percentile[0.25]
        median = percentile[0.50]
        p70 = percentile[0.7]
        p75 = percentile[0.75]
        p80 = percentile[0.8]
        p85 = percentile[0.85]
        p90 = percentile[0.90]
        p95 = percentile[0.95]
        p99 = percentile[0.99]
        max_value = df_subset[j].describe()['max']
        mean_value = np.mean(df_subset[j])
        q_range = p75/p25
        skew = df_subset[j].skew(axis = 0, skipna = True) 
        kurt = df_subset[j].kurt(axis = 0, skipna = True) 
        zero = len(df_subset[df_subset[j] == 0])
        positive = len(df_subset[df_subset[j] > 0])
        negative = len(df_subset[df_subset[j] < 0])
        missing = df_subset[j].isnull().sum()
        temp_list.append((j,coverage,min_value,p1,p5,p10,p25,median,p70,p75,p80,p85,p90,p95,p99,max_value,mean_value,q_range,skew,kurt,zero,positive,negative,missing))

    df_temp = pd.DataFrame(temp_list,columns = ['x','coverage','min','p1','p5','p10','p25','median','p70','p75',
                                                'p80','p85','p90',
                                                'p95','p99','max','mean','q_range','skew','kurt','zero','positive',
                                                'negative','missing'])
    df_temp.to_excel(f"sector_summary_{i}.xlsx")

        
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from scipy.stats import ks_2samp
def logistic(wiki):
    results = dict()
    for i in cols:
        data = wiki[[i, 'margin_change']]
        results[i] = dict()
        # sample split
        data.dropna(inplace=True)
        train, test = train_test_split(data, test_size = 0.3)
        x_train = train.iloc[0:, 0:1]
        y_train = train[['margin_change']]
        x_test = test.iloc[0:, 0:1]
        y_test = test[['margin_change']]
        # fit and predict
        model = LogisticRegression(class_weight="balanced")
        model.fit(x_train, y_train)
        predictions = model.predict(x_test)
        actuals = np.array(y_test)
        # accuracy
        acc = model.score(x_test,actuals)
        results[i]['accuracy'] = acc
        # gini
        fpr, tpr, thresholds = roc_curve(actuals, predictions)
        roc_auc = auc(fpr, tpr)
        gini = (2 * roc_auc) - 1
        results[i]['gini'] = gini
        if gini < 0:
            predictions = 1 - predictions
            fpr, tpr, thresholds = roc_curve(actuals, predictions)
            roc_auc = auc(fpr, tpr)
            gini = (2 * roc_auc) - 1
        results[i]['gini'] = gini
        #k-s statistics
#         print(y_test)
#         print(type(y_test))
#         print(y_test["margin_change"])
#         print(y_test["margin_change"].shape)
#         print(predictions)
# #         predictions.resize((len(predictions), 1))
        ks_result = ks_2samp(y_test["margin_change"],predictions)
        results[i]["ks_stats"] = ks_result[0]
        results[i]["ks_pvalue"] = ks_result[1]
    return results
```
